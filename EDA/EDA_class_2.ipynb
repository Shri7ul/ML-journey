```python
import pandas as pd
import seaborn as sns
```

```python
titanic = pd.read_csv('train.csv')
```

```python
titanic.head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.get_dataset_names()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

## 1. Scatterplot (Numerical - Numerical)

```python
tips.head()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue = tips["sex"], style=tips["smoker"])
```

## 2. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(titanic, x = titanic["Pclass"], y = titanic["Age"], hue = titanic["Sex"])
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(titanic, x = titanic["Pclass"], y = titanic["Age"] , hue=titanic["Survived"])
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```
